```python
def pearson_correlation(df,target_threshold,feature_threshold):
    #df=df.drop(columns='outcome')
    target='outcome'
    corr=df.corr()
    candidates=corr.index[(corr[target].abs()>target_threshold)].values
    candidates=candidates[candidates!=target]
    #print('correlated to',target,': ',candidates)
    removed=[]
    for c1 in candidates:
        for c2 in candidates:
            if (c1 not in removed) and (c2 not in removed):
                if c1 !=c2:
                    coef=corr.loc[c1,c2]
                    if coef>(feature_threshold) or coef<(-feature_threshold):
                        print('coef between {} and {} :{}'.format(c1,c2,coef))
                        removed.append(c1)
    print('removed:', removed)
    return removed
```

# Modelling on Original Dataset ( Logistic Regression and Decision Tree )

```python
#all random from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV 
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score,make_scorer,precision_recall_curve,roc_curve,roc_auc_score,auc
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import StratifiedKFold


def get_best_f1_model_grid_search(model,param_grid,train_features,train_labels):
    skf=StratifiedKFold(n_splits=3,random_state=42,shuffle=True)
    clf=GridSearchCV(estimator=model,param_grid=param_grid,scoring='f1',
                     cv=skf,n_jobs=1)
    clf.fit(train_features,train_labels.values.ravel())
    print('best_score of F1 (from best set of parameteres)',clf.best_score_)
    print('best_param',clf.best_params_)
    print(clf.best_estimator_)
    return clf.best_params_, clf.best_score_

def get_best_auc_model_grid_search(model,param_grid,train_features,train_labels):
    skf=StratifiedKFold(n_splits=3,random_state=42,shuffle=True)
    clf=GridSearchCV(estimator=model,param_grid=param_grid,scoring='roc_auc',
                     cv=skf,n_jobs=1)
    clf.fit(train_features,train_labels.values.ravel())
    print('best_score of auc (from best set of parameteres)',clf.best_score_)
    print('best_param',clf.best_params_)
    print(clf.best_estimator_)
    return clf.best_params_, clf.best_score_
    

#print model performance on each deataset
def print_result(labels,predict_result):
    print('confusion matrix:\n',confusion_matrix(labels,predict_result))
    tn,fp,fn,tp=confusion_matrix(labels,predict_result).ravel()
    print('TN',tn,'\nFP',fp)
    print('FN',fn,'\nTP',tp)
    #print('TN: ', confusion_mtx[0,0],'\tFP:',confusion_mtx[0,1])
    #print('FN: ', confusion_mtx[1,0],'\tTP:',confusion_mtx[1,1])
    print('classification report:\n', classification_report(labels,predict_result))#it wll give text summary of the precision, recall, f1 score for each class
    #print('roc score',metrics.roc_auc_score(labels,predict_result))#
    print('f1 score on test set:\n',metrics.f1_score(labels,predict_result))
    
    print('precision = positive predictive value = (TP/(TP+FP)) on test set\n',metrics.precision_score(labels,predict_result))
    
    print('recall = sensitivity = tpr = (TP/(TP+FN)) on test set\n',metrics.recall_score(labels,predict_result))
    
    print('specificity = (TN/(TN+FP)) on test set\n',(tn/(tn+fp)))
    
    print('fpr = 1-specificity = (FP/(TN+FP)) on test set\n',(fp/(tn+fp)))
    
    print('negative predictive value- NPV (TN/(TN+FN)) on test set\n',(tn/(tn+fn)))
    
    print('roc_auc on test set\n',metrics.roc_auc_score(labels,predict_result))
    
def predict_probabilities(model, test_features):
    predicted_probs = model.predict_proba(test_features)
    return predicted_probs
    
def plot_roc_auc(model,model_name,test_features, test_labels):
    # generate a no skill prediction (majority class)
    ns_probs = [0 for _ in range(len(test_labels))]
    pre_probs = predict_probabilities(model, test_features)
    # keep probabilities for the positive outcome only
    pre_probs = pre_probs[:, 1]
    # calculate scores
    ns_auc = roc_auc_score(test_labels, ns_probs)
    model_auc = roc_auc_score(test_labels, pre_probs)
    # summarize scores
    print('No Skill: ROC AUC=%.3f' % (ns_auc))
    print('Logistic: ROC AUC=%.3f' % (model_auc))
    # calculate roc curves
    ns_fpr, ns_tpr, _ = roc_curve(test_labels, ns_probs)
    model_fpr, model_tpr, _ = roc_curve(test_labels, pre_probs)
    # plot the roc curve for the model
    plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
    plt.plot(model_fpr, model_tpr, marker='.', label=str(model_name))
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()

def plot_precision_recall(model, model_name,test_features, test_labels):
    # predict probabilities
    pre_probs = predict_probabilities(model, test_features)
    # keep probabilities for the positive outcome only
    pre_probs = pre_probs[:, 1]
    # predict class values
    yhat = model.predict(test_features)
    model_precision, model_recall, _ = precision_recall_curve(test_labels, pre_probs)
    model_f1, model_auc = f1_score(test_labels, yhat), auc(model_recall, model_precision)
    # summarize scores
    print('Logistic: f1=%.3f auc=%.3f' % (model_f1, model_auc))
    # plot the precision-recall curves
    no_skill = len(test_labels[test_labels==1]) / len(test_labels)
    plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    plt.plot(model_recall, model_precision, marker='.', label=str(model_name))
    # axis labels
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()
```

# SMOTE to balance dataset

```python
## In many business cases, false positives are less problematic than false negatives.
#Using SMOTE we can tweak the model to reduce false negatives, at the cost of increasing false positives. 
#The result of using SMOTE is generally an increase in recall, at the cost of lower precision. 
#This means that we will add more predictions of the minority class: some of them correct (increasing recall), but some of them wrong (decreasing precision).

#SMOTE increases recall at the cost of lower precision

from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(train.drop(columns='outcome'), train['outcome'])

pd.Series(y_resampled).value_counts().plot(kind='bar', title='Class distribution after appying SMOTE', 
                                           xlabel='outcome')
```
